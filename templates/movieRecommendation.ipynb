```python
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
def genresconvert(obj):
    l =[]
    for i in ast.literal_eval(obj):
        l.append(i['name'])
    return l
```

```python
def director(obj):
    l =[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            l.append(i['name']) 
    return l
```

```python
def RemoveSpace(obj):
    l=[]
    for i in obj:
        l.append(i.replace(" ", ""))
    return l
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
stem ('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. SamWorthington ZoeSaldana SigourneyWeaver StephenLang culture clash future space war space colony society space travel futuristic romance space alien tribe alien planet cgi marine soldier battle love affair anti war power relations mind and soul 3d SamWorthington ZoeSaldana SigourneyWeaver StephenLang SamWorthington ZoeSaldana SigourneyWeaver StephenLang')
```

```python
new['tags']=new['tags'].apply(stem)
```

#### counter vectorization

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features = 5000, stop_words='english')
```

```python
vector = cv.fit_transform(new['tags']).toarray()
```

```python
from sklearn.metrics. pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vector)
```
